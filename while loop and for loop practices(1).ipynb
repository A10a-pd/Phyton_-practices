1) take the number of lessons and their marks from user and calculate the average.if avg<10:print fail,if 10<=avg<14: print rank3,if 14<=avg<17:print rank2,if 17<=avg<=20:print rank1,otherwise,print invalid. 

solution using "while loop":

```python
n=int(input("enter the number ofmbers of lessons"))
i=1
s=0
while i<=n:
    mark=float(input("enter your mark"))
    s+=mark
    i+=1
avg=s/n
print(avg)
if avg<10:
    print("fail")
elif 10<=avg<14:
    print("rank3")
elif 14<=avg<17:
    print("rank2")
elif 17<=avg<=20:
    print("rank1")
else:
    print("invalid")
```

solution using "for loop":

```python
n=int(input("enter the number of lessons"))
s=0
for i in range(1,n+1):
    mark=float(input("enter your mark"))
    s+=mark
    avg=s/n
print(avg)
if avg<10:
    print("fail")
elif 10<=avg<14:
    print("rank3")
elif 14<=avg<17:
    print("rank2")
elif 17<=avg<=20:
    print("rank1")
else:
    print("invalid")
```

2) calculate this:  1/1!+1/2!+1/3!+....+1/n!

solution using "while loop":

```python
n=int(input("enter a number"))
i=1
f=1
s=0
while i<=n:
    f=i*f
    s=s+1/f
    i+=1
print(s)    
```

```python
n=int(input("enter a number:"))
f=1
s=0
for i in range(1,n+1):
    f=i*f
    s=s+1/f
print(s)
```

8. write a program that shows all common divisors of 2 numbers

 first method. 

```python
a=int(input("enter a number"))
b=int(input("enter a number"))
if a<b:
    smaller=a
else:
    smaller=b
for i in range(1,smaller+1):
    if (a%i==0) and (b%i==0):
        print(i )   
```
